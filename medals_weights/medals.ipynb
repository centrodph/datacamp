```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

summer_2016_medals = pd.read_csv('./summer2016.csv')

print(summer_2016_medals.head())
```

```python
fig, ax = plt.subplots()

sports = summer_2016_medals["Sport"].unique()

for sport in sports:
  sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sport]
  ax.bar(sport, sport_df["Weight"].mean(), yerr= sport_df["Weight"].std())

ax.set_ylabel("Weight")
ax.set_xticklabels(sports, rotation=90)
fig.set_size_inches([10, 4])
plt.show()
```

```python
fig, ax = plt.subplots()

sorted_df = pd.DataFrame({'Sport': sports })
sorted_df['Mean'] = 0
sorted_df['Std'] = 0

for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sport]
    sorted_df.loc[sorted_df.Sport == sport, "Mean"] = sport_df["Weight"].mean()
    sorted_df.loc[sorted_df.Sport == sport, "Std"] = sport_df["Weight"].std()    

sorted_df  = sorted_df.sort_values("Mean")    

for sport in sorted_df.Sport:
    ax.bar(sport, sorted_df[sorted_df["Sport"] == sport]["Mean"], yerr= sorted_df[sorted_df["Sport"] == sport]["Std"])

ax.set_ylabel("Weight")
ax.set_xticklabels(sports, rotation=90)
fig.set_size_inches([10, 4])
plt.show()
```
